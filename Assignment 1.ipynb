##  Checking Outliers in my numerical features using the boxplot technique

```python

plt.figure(figsize = (15,15), facecolor = 'white')
plotnumber = 1
for column in numerical_fetures.iloc[:,0:10]:
    plt.subplot(5,2,plotnumber)
    sns.boxplot(numerical_fetures[column])
    plt.ylabel(column,fontsize = 20)
    plotnumber+=1  
    
plt.tight_layout()
plt.show()
```

#### We can see from the boxplot , as per the data description in the pdf file X can have values between  1 to 9  but as seen in the boxplot  there are some values which are outside the range.
#### Similarly ,  in the RH column values should be in the range of  0 to 100 but from the boxplot it is seen that value are outside this range.
#### For column A also there are values which are outside the  given range of  18.7 to 96.20 .

```python
## Checking the outliers in the X column with the help of data description given in the file.
df[(df["X"]>9)|(df["X"]<1)]
```

```python
## Checking the outliers in the RH column
df[(df["RH"]>100)|(df["RH"]<15)]
```

```python
## Checking the outliers in the A column
df[(df["A"]>96.20)|(df["A"]<18.7)]
```

```python
## Finding the boundary values
for columns in numerical_fetures.iloc[:,0:10]:
    print(columns)
    print(f"Highest Value is {numerical_fetures[columns].mean() + 3*numerical_fetures[columns].std()}")
    print(f"Lowest Value is {numerical_fetures[columns].mean() - 3*numerical_fetures[columns].std()}")
    print("")
    
```

```python
# Define the conditions for outliers in each column
X_condition = (df["X"] > 9) | (df["X"] < 1)
Rh_condition = (df["RH"] > 100) | (df["RH"] < 15)
A_condition = (df["A"] > 96.20) | (df["A"] < 18.7)

# Combine the conditions using logical OR (|) to identify rows with outliers
outlier_condition = X_condition | Rh_condition | A_condition

# Create a new DataFrame without the outliers
df_no_outliers = df[~outlier_condition]

# Now df_no_outliers contains your DataFrame with outliers removed
```

#### Lets Proceed with classification approach
#### We need to convert the target variable into binary form that is area which is not affected is "0" and the affected area is "1".
#### After binarizing the target variable , we can use the Classification Method.

```python
# Creating a classification model
from sklearn.linear_model import LogisticRegression
```

```python
# Convert the target variable 'area' to binary
y_binary = (y > 0).astype(int)
```

```python
# Split the data into training (80%) and testing (20%) sets with the binary target
X_train_Scaled, X_test_Scaled, y_train_bin, y_test_bin = train_test_split(X, y_binary, test_size=0.2, random_state= 0)
```

```python
# Check the distribution of the binary target
y_binary.value_counts()
```

### The code trains a logistic regression model on X_train and evaluates its performance on X_test using accuracy and other metrics. The final accuracy is then printed as a percentage

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Initializing the logistic regression model
logistic = LogisticRegression(max_iter=20000, random_state= 0)
```

```python
# Model fitting into  training data
logistic.fit(X_train, y_train_bin)
```

```python
# Predicting  the binary target variable for the testing set
y_pred_bin = logistic.predict(X_test)
```

```python
# Performance metrics
Accuracy = accuracy_score(y_test_bin, y_pred_bin)
Classification_rep = classification_report(y_test_bin, y_pred_bin)

Accuracy, Classification_rep

print(f"Accuracy using Logistic Regression: {Accuracy * 100}%")
print(Classification_rep)
```
